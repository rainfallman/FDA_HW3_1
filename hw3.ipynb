```python
import numpy as np
import pandas as pd

df = pd.read_csv('sp500t.csv')     
tf=pd.read_csv('sp500te.csv') 
```

```python
#train資料的前處理
print(df)
tmp=df["Close Price"]-df["Open Price"]


z=np.zeros(2264,dtype=np.int)
t_dif=np.zeros(2264,dtype=np.float)


for i in range(tmp.size):
  if i>0:
    t_dif[i]=df["Open Price"][i]-df["Close Price"][i-1]
  if tmp[i] >=0:
    z[i]=1
  else:
    z[i]=0

print(t_dif)
train_y=pd.DataFrame({'price':z})
train_x=pd.DataFrame({
    'Hd':df["High Price"]-df["Open Price"],
    'Ld':df["Open Price"]-df["Low Price"]
    })

#test資料的前處理

tmp=tf["Close Price"]-tf["Open Price"]
print(tmp)

z=np.zeros(232,dtype=np.int)
te_dif=np.zeros(232,dtype=np.float)
for i in range(tmp.size):
  if i>0:
    te_dif[i]=tf["Open Price"][i]-tf["Close Price"][i-1]
  if tmp[i] >=0:
    z[i]=1
  else:
    z[i]=0


test_y=pd.DataFrame({'price':z})
test_x=pd.DataFrame({
    'Hd':tf["High Price"]-tf["Open Price"],
    'Ld':tf["Open Price"]-tf["Low Price"]
    })

#額外資料的前處理
t_op=df["Open Price"]
te_op=tf["Open Price"]
t_v=df[ "Volume"]/10000000
te_v=tf[ "Volume"]/10000000
```

LogisticRegression part


```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python

clf=LogisticRegression(random_state=0).fit(train_x, train_y)
prey=clf.predict(test_x)
acc = accuracy_score(test_y, prey)
print(acc)
```

```python
#加入當天開始的開盤價格

train_x1=train_x
train_x1["OP"]=t_op

test_x1=test_x
test_x1["OP"]=te_op

clf=LogisticRegression(random_state=0).fit(train_x1, train_y)
prey=clf.predict(test_x1)
acc = accuracy_score(test_y, prey)
print(acc)
```

```python
#加入當天的交易量

train_x1=train_x
train_x1["V"]=t_v

test_x1=test_x
test_x1["V"]=te_v

clf=LogisticRegression(random_state=0).fit(train_x1, train_y)
prey=clf.predict(test_x1)
acc = accuracy_score(test_y, prey)
print(acc)
```
