這次我的第三個 classifiers 我選擇使用svm來

然後有關於資料前處理的部分，我將當天的最高價去扣當天的開盤價格，還有當天開盤價去扣當天的最低價，這兩筆當作一定要輸入的，我認為判斷當天是上漲或下跌這很重要，然後我有額外把當天的開盤價，還有就是把交易量除以1百萬，還有就是用今天的開盤價去減昨天的收盤價當作其他可變的訓練參數

所以我總共有2個一定會訓練的參數，3個可變的參數  
固定：最高價去扣當天的開盤價 -最高價差  

固定：開盤價去扣當天的最低價 -最低價差

可變：今天的開盤價  
可變：交易量除以1百萬   
可變：今天的開盤價去減昨天的收盤價 -開盤走勢  

然後在只有兩個一定的訓練下，表現最好的是svm ，logistic 跟NN 表現一樣

但是我覺得NN的表現應該要更好，所以我嘗試了很多不同的hidden layer想要提升NN的表現，因為對NN來說hidden layer影響很大，其他的classifiers做修改效果也沒有很好，後來layer為(8,4)的有成功衝到0.849，不過我覺得應該可以弄得更好，只是我找不到更好的hidden layer

然後當開始丟其他可變參數的時候，我發現當天的開盤價格不會影響到任何一個表現，我猜是開盤價與上漲下跌沒有關係

而對於NN來說，開盤價格不會受到影響，當天交易量也不會受到影響，可是NN卻會被當天開盤走勢所影響，而且表現還變差了，而且在丟入走勢的情況下，如果再丟入交易量，會讓NN表現下降更多

但是對於SVM來說，丟入開盤走勢會讓SVM表現上升，而丟入走勢則會下降，最好的表現就是只丟入開盤走勢，可以提升到大概0.84
